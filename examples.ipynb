# Examples

## Compute splicing dependency

```python
import os
from target_spotter import SplicingDependency
```

```python
ROOT = "~/projects/publication_splicing_dependency"
PREP_DIR = os.path.join(ROOT, "data", "prep")
RESULTS_DIR = os.path.join(ROOT, "results", "model_splicing_dependency")
splicing_file = os.path.join(PREP_DIR, "event_psi", "CCLE-INT.tsv.gz")
genexpr_file = os.path.join(PREP_DIR, "genexpr_tpm", "CCLE.tsv.gz")
coefs_dir = os.path.join(RESULTS_DIR, "files", "models_gene_dependency-INT")
```

```python
estimator = SplicingDependency.PredictFromFiles(
    splicing_file,
    genexpr_file,
    coefs_splicing_file=os.path.join(coefs_dir, "coefs_splicing.pickle.gz"),
    coefs_genexpr_file=os.path.join(coefs_dir, "coefs_genexpr.pickle.gz"),
    coefs_interaction_file=os.path.join(coefs_dir, "coefs_interaction.pickle.gz"),
    coefs_intercept_file=os.path.join(coefs_dir, "coefs_intercept.pickle.gz"),
)
estimator.load_data()
```

```python
model = SplicingDependency.SplicingDependency(
    normalize_counts=False, n_iterations=2, n_jobs=10
)
model.predict(
    estimator.splicing_.iloc[:100],
    estimator.genexpr_,
    estimator.ccle_stats_,
    estimator.coefs_splicing_,
    estimator.coefs_genexpr_,
    estimator.coefs_interaction_,
    estimator.coefs_intercept_,
)
```
